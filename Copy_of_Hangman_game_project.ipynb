```
import random

list_of_words = ['director', 'minister','governor', 'chairman']
selected = random.choice(list_of_words)

name = input('kindly enter your name?: ')
print('welcome to hangman game project:', name)

print('the word to guess is', len(selected), 'letters word')
right_guess = ['_'] * len(selected)
wrong_guess = []

def printletters():
    for letter in right_guess:
        print(letter.upper(), end=' ')
    print()

printletters()

def draw_hangman(number_of_tries):
    drawing = ['''                 
    __________
    |
    |
    |
    |
    |
____|_____''', '''
    __________
    |        |
    |        
    |       
    |       
    |      
____|_____''', '''
    __________
    |        |
    |        O
    | 
    |
    |   
____|_____''', '''
    __________
    |        |
    |        O
    |        |
    |        
    |       
____|_____''', '''
    __________
    |        |
    |        O
    |        |
    |        |
    |   
____|_____''', '''
    __________
    |        |
    |        O
    |       /|
    |        |
    |     
____|_____''', '''
    __________
    |        |
    |        O
    |       /|\\
    |        |
    |      
____|_____''', '''
    __________
    |        |
    |        O
    |       /|\\
    |        |
    |       / 
____|_____''', '''
    __________
    |        |
    |        O
    |       /|\\
    |        |
    |       / \\
____|_____''']
    return drawing[number_of_tries]

play='y'
while play == 'y':
    counter = 0
    selected = random.choice(list_of_words)
    right_guess = ['_'] * len(selected)
    wrong_guess = []

    while True:
            print(draw_hangman(len(wrong_guess)))
            guess = input('pick a letter: ')

            if guess.isalpha() and len(guess) == 1:
                if guess in selected:
                    index = 0
                    for letter in selected:
                        if letter == guess:
                            right_guess[index] = guess
                        index = index + 1
                    printletters()
                else:
                    if guess not in wrong_guess:
                        counter = counter + 1
                        wrong_guess.append(guess)
                    else:
                        print('you already attempted that')
                        print(wrong_guess)
                if '_' not in right_guess:
                    print('you win')
                    break
                if counter >= 8:
                    print('you just lost')
                    print('The correct word was ',selected.upper())
                    print('your wrong letters were : ', wrong_guess)
                    print(draw_hangman(len(wrong_guess)))
                    break
            else:
                print("'please enter a letter")
    play=input('do you want to play again? (y/n):')
print('goodbye')
```
